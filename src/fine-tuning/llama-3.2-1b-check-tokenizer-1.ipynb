```python
# difine tokenizer
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.2-1B")
```

```python
dataset_max_length = 512
PAD_TOKEN = "<|end_of_text|>"
```

```python
# print meta-llama/Llama-3.2-1B tokenizer special tokens
import json
import pandas

with open("./llama-3.2-1b_tokenizer.json", 'r', encoding='utf-8') as f:
    added_tokens_df = pandas.DataFrame(json.load(f)['added_tokens'])

# all print setting
pandas.set_option('display.max_rows', None)
pandas.set_option('display.max_columns', None)
pandas.set_option('display.width', None)
pandas.set_option('display.max_colwidth', None)

print(added_tokens_df)
```

```python
# formatted text
formatted_text = [
 "<|begin_of_text|><|start_header_id|>user<|end_header_id|>Give three tips for staying healthy.<|eot_id|><|start_header_id|>assistant<|end_header_id|>1.Eat a balanced diet and make sure to include plenty of fruits and vegetables. 2. Exercise regularly to keep your body active and strong. 3. Get enough sleep and maintain a consistent sleep schedule.<|eot_id|><|end_of_text|>", 
 "<|begin_of_text|><|start_header_id|>user<|end_header_id|>What are the three primary colors?<|eot_id|><|start_header_id|>assistant<|end_header_id|>The three primary colors are red, blue, and yellow.<|eot_id|><|end_of_text|>"
]
```

```python
# print formatted text list
input_word_list = []

for i in range(len(formatted_text)):
    input_word = formatted_text[i]
    input_word_list.append(input_word)
    print(input_word_list)
```

```python
# encode formatted text
input_id_list = []

for i in range(len(formatted_text)):
    input_id = tokenizer.encode(formatted_text[i])
    print(input_id)
    input_id_list.append(input_id)
```

```python
# token prediction (encode version)
encode_sample = input_id_list[0][50:]

context_size = 10
x = encode_sample[:context_size]
y = encode_sample[1:context_size+1]

print(f"x: {x}")
print(f"y:       {y}")
```

```python
# token prediction (encode version)

for i in range(1, context_size+1):
    context = encode_sample[:i]
    desired = encode_sample[i]
    print(context, "---->", desired)
```

```python
# decode input_ids
for i in range(len(input_id_list)):
    print(tokenizer.decode(input_id_list[i]))
```
